### 5. Replicando os resultados do capítulo

```python
# Removendo colunas que não serão utilizadas
dados = dados.drop(columns=['eligible', 'treatment_locality', 'enrolled'])
dados.head()
```

```python
# Filtrar para as duas rodadas
baseline_data = dados[dados['round'] == 0]
followup_data = dados[dados['round'] == 1]

# Inicializar tabela
results = []

# Variável: Despesas com saúde na linha de base
group1 = baseline_data[baseline_data['promotion_locality'] == 1]['health_expenditures']
group2 = baseline_data[baseline_data['promotion_locality'] == 0]['health_expenditures']
mean1, mean2 = group1.mean(), group2.mean()
diff = mean1 - mean2
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
results.append(['Despesas com saúde (linha de base)', f"{mean1:.2f}", f"{mean2:.2f}", f"{diff:.2f}", f"{t_stat:.2f}", f"{p_value:.4f}"])

# Variável: Despesas com saúde no acompanhamento
group1 = followup_data[followup_data['promotion_locality'] == 1]['health_expenditures']
group2 = followup_data[followup_data['promotion_locality'] == 0]['health_expenditures']
mean1, mean2 = group1.mean(), group2.mean()
diff = mean1 - mean2
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
results.append(['Despesas com saúde (acompanhamento)', f"{mean1:.2f}", f"{mean2:.2f}", f"{diff:.2f}", f"{t_stat:.2f}", f"{p_value:.4f}"])

# Variável: Taxa de inscrição
group1 = followup_data[followup_data['promotion_locality'] == 1]['enrolled_rp']
group2 = followup_data[followup_data['promotion_locality'] == 0]['enrolled_rp']
mean1, mean2 = group1.mean() * 100, group2.mean() * 100  # Transformar em porcentagem
diff = mean1 - mean2
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
results.append(['Taxa de inscrição no HISP', f"{mean1:.2f}%", f"{mean2:.2f}%", f"{diff:.2f}%", f"{t_stat:.2f}", f"{p_value:.4f}"])
```

```python
# Exibir a tabela formatada
headers = ['Descrição', 'Média Encorajados', 'Média Não Encorajados', 'Diferença', 'Estatística t', 'p-valor']
print(tabulate(results, headers=headers))
```
